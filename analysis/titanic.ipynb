```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# Задание 1
data_titanic = pd.read_csv('.\\data\\dataset_titanic.csv', header=0)


# Задание 2
print(data_titanic.size)
print(data_titanic.shape)
print(data_titanic.columns)
for i in data_titanic.columns:
    print(f'Уникальные значения из столбца {i}:\n{data_titanic[i].unique()}')
for i in data_titanic.columns:
    print(f'Количество уникальных значений столбца {i}:\n{len(data_titanic[i].unique())}')
```

```python
# Задание 3
new_data_titanic = data_titanic.copy()
new_data_titanic.pop('Name')
new_data_titanic.pop('Sex')
new_data_titanic = new_data_titanic[new_data_titanic['Survived']==1]
new_data_titanic = new_data_titanic.sort_values(by=['Age'])




# Задание 4
new_data_titanic.to_csv('new_data_titanic.txt', sep=',')




# Задание 5
all_passengers = data_titanic
print(all_passengers['Age'].median())

survived = data_titanic[data_titanic['Survived']==1]
print(survived['Age'].median())

not_survived = data_titanic[data_titanic['Survived']==0]
print(not_survived['Age'].median())

male = data_titanic[data_titanic['Sex']=='male']
print(male['Age'].median())

female = data_titanic[data_titanic['Sex']=='female']
print(female['Age'].median())

survived_male = survived[survived['Sex']=='male']
print(survived_male['Age'].median())

survived_female = survived[survived['Sex']=='female']
print(survived_female['Age'].median())

not_survived_male = not_survived[not_survived['Sex']=='male']
print(not_survived_male['Age'].median())

not_survived_female = not_survived[not_survived['Sex']=='female']
print(not_survived_female['Age'].median())
```

```python
# Задание 6
fig1 = plt.figure()
axs1 = fig1.add_subplot(1, 2, 1, title='male', xlabel='Fare', ylabel='person')
gph1 = axs1.hist(data_titanic[data_titanic['Sex']=='male']['Fare'], 25)
axs2 = fig1.add_subplot(1, 2, 2, title='female', xlabel='Fare', ylabel='person')
gph2 = axs2.hist(data_titanic[data_titanic['Sex']=='female']['Fare'], 25)
plt.show()
```

```python
# Задание 7
fig1 = plt.figure()
axs1 = fig1.add_subplot(1, 1, 1, xlabel='Class', ylabel='Fare')
gph1 = axs1.scatter(data_titanic['Pclass'], data_titanic['Fare'])
plt.show()
```

```python
# Задание 8
pclass = data_titanic['Pclass'].unique()

survived = data_titanic[data_titanic['Survived']==1]
survived_class_1 = len(survived[survived['Pclass']==1])
survived_class_2 = len(survived[survived['Pclass']==2])
survived_class_3 = len(survived[survived['Pclass']==3])

not_survived = data_titanic[data_titanic['Survived']==0]
not_survived_class_1 = len(not_survived[not_survived['Pclass']==1])
not_survived_class_2 = len(not_survived[not_survived['Pclass']==2])
not_survived_class_3 = len(not_survived[not_survived['Pclass']==3])

survived_classes = [survived_class_1, survived_class_2, survived_class_3]
not_survived_classes = [not_survived_class_1, not_survived_class_2, not_survived_class_3]

fig1 = plt.figure()
axs1 = fig1.add_subplot(1, 1, 1, xlabel='Class', ylabel='person')
gph1 = axs1.bar(pclass-0.1, survived_classes, width=0.3, label='survived')
gph2 = axs1.bar(pclass+0.1,  not_survived_classes, width=0.3, label='not survived')
axs1.legend()
plt.show()
```

```python
# Задание 9
fig1 = plt.figure()
axs1 = fig1.add_subplot(1, 2, 1, title='survived', xlabel='Age', ylabel='person')
gph1 = axs1.hist(data_titanic[data_titanic['Survived']==1]['Age'], 25)
axs2 = fig1.add_subplot(1, 2, 2, title='not_survived', xlabel='Age', ylabel='person')
gph2 = axs2.hist(data_titanic[data_titanic['Survived']==0]['Age'], 25)
plt.show()
```
