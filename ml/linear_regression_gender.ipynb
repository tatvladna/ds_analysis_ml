```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# Задание 1

```python
data = pd.read_csv('.\\data\\dataset_points.csv')
# преобразование в двухмерный массив
x = np.reshape(data['0'], [-1, 1])
y = np.reshape(data['1'], [-1, 1])
```

```python
model_f = LinearRegression()
model_f.fit(x, y)
```

```python
k = model_f.coef_ # тангенс угла наклона = k, угловой коэффициент
b = model_f.intercept_  #  точка пересечения линии регрессии с oy

x_line = x
y_line = k * x + b # линия регрессии
```

```python
# линейная регрессия, которая лучше всего соответствует исходным данным по методу наименьших квадратов.
fig1 = plt.figure()
axs1 = fig1.add_subplot(1, 1, 1, xlabel='x', ylabel='y')
gph1 = axs1.scatter(x, y)
gph2 = axs1.plot(x_line, y_line, color='r')
plt.show()
```

# Задание 2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
# Линейная регрессия на ирисах
x0, y0 = load_iris(return_X_y=True)
x = x0[:, 0:3]
y = x0[:, 3]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

model_F = LinearRegression()
model_F.fit(x_train, y_train)
y_pred = model_F.predict(x_test)

MAE = metrics.mean_absolute_error(y_pred, y_test)
RMSE = metrics.root_mean_squared_error(y_pred, y_test)
display(f'MAE = {MAE}')
display(f'RMSE = {RMSE}')
display(f'MAE_percent = {MAE / np.mean(y_test)* 100} %')
display(f'RMSE_percent = {RMSE / np.mean(y_test)* 100} %')
```

# Задание 3

```python
# Задание 3

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

data = pd.read_csv('.\\data\\dataset_gender.csv')
x = data[['Height', 'Weight']]
y = data['Gender']
y = np.array([1 if i == 'Male' else 0 for i in y])

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1)

model_F = KNeighborsClassifier(n_neighbors=1000)
model_F.fit(x_train, y_train)
y_pred = model_F.predict(x_test)

fig1 = plt.figure()
axs1 = fig1.add_subplot(1, 1, 1, xlabel='Height', ylabel='Weight')
gph1 = axs1.scatter(x_train['Height'], x_train['Weight'], c=y_train)
gph2 = axs1.scatter(x_test['Height'], x_test['Weight'], color=[0.5, 0.5, 0.5])
plt.show()

display(f'Точность модели = {metrics.accuracy_score(y_pred, y_test)}')
```

```python
# Задание 4

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import LinearSVC
from sklearn.preprocessing import Normalizer

data = pd.read_csv('.\\data\\dataset_class.csv')
x = data[['0', '1']]
x1 = data['0']
x2 = data['1']
y = data['2']

# Создание и обучение модели линейного классификатора
model_F = LinearSVC()
model_F.fit(x, y)

# получение коэффициентов модели
w = model_F.coef_

# построение разделяющей линии
x_line = np.arange(-10, 20)  # создание диапазона значений для x
y_line = - x_line * w[0, 0] / w[0, 1]  # вычисление y значений для разделяющей линии

fig1 = plt.figure()
axs1 = fig1.add_subplot(1, 1, 1, xlabel='x1', ylabel='x2')
gph1 = axs1.scatter(x1, x2, c=y)
ghp2 = axs1.plot(x_line, y_line)
plt.show()
```
